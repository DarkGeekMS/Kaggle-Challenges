```python
import os 
import cv2
import matplotlib.pyplot as plt
import pydicom
from PIL import Image
```

```python
files = os.listdir("stage_2_images")
```

```python
files
```

```python
for file in files:
    img = pydicom.dcmread("stage_2_images/"+file).pixel_array
    img = Image.fromarray(img).convert("L")
    plt.imsave("test2_png/"+file[:-4]+".png", img)
```
